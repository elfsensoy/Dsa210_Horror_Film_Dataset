```python
import pandas as pd

# IMBD Movie Data Cleaning
movies_raw = pd.read_csv("../data_raw/Horror Movies IMDb.csv")

movies = movies_raw.rename(columns={
    "Movie Title": "title",
    "Movie Year": "year",
    "Runtime": "runtime_min",
    "Genre": "genre",
    "Rating": "imdb_rating",
    "Director": "director",
    "Votes": "votes"
})

movies["year"] = pd.to_numeric(movies["year"], errors="coerce")
movies["runtime_min"] = pd.to_numeric(movies["runtime_min"], errors="coerce")
movies["imdb_rating"] = pd.to_numeric(movies["imdb_rating"], errors="coerce")

movies["votes"] = (
    movies["votes"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .astype("Int64")
)

movies = movies.dropna(subset=["title", "year", "imdb_rating"])
movies.to_csv("../data_clean/movies_clean.csv", index=False)
```

```python
# TMBD Reviews Data Cleaning
reviews_tmdb_raw = pd.read_csv("../data_clean/horror_tmdb_reviews_sample.csv")

reviews_tmdb = reviews_tmdb_raw.rename(columns={
    "content": "review_text",
    "author": "author",
    "author_rating": "rating",
    "movie_title": "title",
    "movie_year": "year"
})

reviews_tmdb["year"] = pd.to_numeric(reviews_tmdb["year"], errors="coerce")
reviews_tmdb["rating"] = pd.to_numeric(reviews_tmdb["rating"], errors="coerce")

reviews_tmdb.to_csv("../data_clean/reviews_tmdb_clean.csv", index=False)
```
